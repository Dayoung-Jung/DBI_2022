# **data set**

```python
df_info = pd.read_csv('/content/drive/Shareddrives/2022 디지털산업혁신 빅데이터 플랫폼 아이디어 경진대회/YG/processed_dataset/사업자등록번호-산업분류코드(90427개).csv')
df_info = df_info[['사업자등록번호', '산업분류코드', '중분류_산업분류코드', '대분류_산업분류코드']]
df_info.head()
```

```python
df_tmp = pd.read_csv('/content/drive/Shareddrives/2022 디지털산업혁신 빅데이터 플랫폼 아이디어 경진대회/JDY/result/df_final.csv')
df_tmp.head()
```

```python
df_merge.columns
```

# **자기성장성 등급화**

```python
df_cluster = df_merge[['자기성장성']]
df_cluster
```

```python
ks = range(1,10)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(df_cluster)
    inertias.append(model.inertia_)

plt.figure(figsize=(4, 4))

plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
clust_model = KMeans(n_clusters = 5 
                     , n_init=100 
                     , max_iter=500 
                     , random_state = 42
                     , algorithm='auto'
                    )

clust_model.fit(df_cluster)  

centers = clust_model.cluster_centers_ 
pred = clust_model.predict(df_cluster) 

pd.DataFrame(centers)
```

```python
df_cluster['clust'] = pred
df_cluster.loc[df_cluster['clust'] == 1, '자기성장성_등급'] = 'A'
df_cluster.loc[df_cluster['clust'] == 4, '자기성장성_등급'] = 'B'
df_cluster.loc[df_cluster['clust'] == 2, '자기성장성_등급'] = 'C'
df_cluster.loc[df_cluster['clust'] == 0, '자기성장성_등급'] = 'D'
df_cluster.loc[df_cluster['clust'] == 3, '자기성장성_등급'] = 'E'

df_cluster = df_cluster.drop(['clust'], axis = 1)

df_cluster
```

```python
df_merge['자기성장성_등급'] = df_cluster['자기성장성_등급']
df_merge.head()
```
