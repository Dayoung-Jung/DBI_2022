# **DataSet 재구성**

```python
df_tmp = df.copy()

df_tmp['대분류명'] = df_tmp['대분류명'].str.replace(pat = r'[^\w]', repl = r' ', regex = True)
df_tmp['대분류명'] = df_tmp['대분류명'].str.replace(r"[0-9]", "")
df_tmp['전화번호'] = df_tmp['전화번호'].str.replace(' ', '-')
df_tmp['전화번호'] = df_tmp['전화번호'].str.replace('--', '-')


df_tmp = df_tmp.replace('알수없음(1)', '-')
df_tmp = df_tmp.replace('알수없음(2)', '-')



df_tmp.head()
```
