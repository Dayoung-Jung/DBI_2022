# **근속기간 (중분류)**

```python
df_tmp = pd.read_csv('/content/drive/Shareddrives/2022 디지털산업혁신 빅데이터 플랫폼 아이디어 경진대회/dataset/기업성장_16종/기업정보상세.csv')
df_period = df_tmp.copy()

df_period = df_period[['사업자등록번호', '산업분류코드', '종업원수']]
df_period['산업분류코드'] = df_period['산업분류코드'].fillna(00)


df_period.loc[df_period['종업원수'] >= 500,'종업원_평균근속기간'] = 11.2
df_period.loc[df_period['종업원수'] <= 499,'종업원_평균근속기간'] = 8.7
df_period.loc[df_period['종업원수'] <= 299,'종업원_평균근속기간'] = 7.5
df_period.loc[df_period['종업원수'] <= 99,'종업원_평균근속기간'] = 6.4
df_period.loc[df_period['종업원수'] <= 29,'종업원_평균근속기간'] = 5.7
df_period.loc[df_period['종업원수'] <= 9,'종업원_평균근속기간'] = 5.6

# df_period['종업원_평균근속기간'] = df_period['종업원_평균근속기간'].fillna(7.0)
df_period = df_period[['사업자등록번호', '산업분류코드', '종업원_평균근속기간']]


df_period_v2 = df_tmp.copy()

df_period_v2 = df_period_v2[['사업자등록번호', '산업분류코드']]
df_period_v2['산업분류코드'] = df_period_v2['산업분류코드'].fillna(00)

df_period_v2['산업분류코드'] = df_period_v2['산업분류코드'].astype('str')
df_period_v2['산업분류코드'] = df_period_v2['산업분류코드'].str[:2]
df_period_v2['산업분류코드'] = df_period_v2['산업분류코드'].astype('float')

df_period_v2.loc[df_period_v2['산업분류코드'] >= 94,'산업별_평균근속기간'] = 5.4
df_period_v2.loc[df_period_v2['산업분류코드'] <= 93,'산업별_평균근속기간'] = 5.6
df_period_v2.loc[df_period_v2['산업분류코드'] <= 89,'산업별_평균근속기간'] = 4.6
df_period_v2.loc[df_period_v2['산업분류코드'] <= 85,'산업별_평균근속기간'] = 8.6
df_period_v2.loc[df_period_v2['산업분류코드'] <= 84,'산업별_평균근속기간'] = 5.1 
df_period_v2.loc[df_period_v2['산업분류코드'] <= 73,'산업별_평균근속기간'] = 6.9
df_period_v2.loc[df_period_v2['산업분류코드'] <= 69,'산업별_평균근속기간'] = 4.8
df_period_v2.loc[df_period_v2['산업분류코드'] <= 66,'산업별_평균근속기간'] = 11.6
df_period_v2.loc[df_period_v2['산업분류코드'] <= 63,'산업별_평균근속기간'] = 6.6
df_period_v2.loc[df_period_v2['산업분류코드'] <= 56,'산업별_평균근속기간'] = 3.4
df_period_v2.loc[df_period_v2['산업분류코드'] <= 52,'산업별_평균근속기간'] = 8.2
df_period_v2.loc[df_period_v2['산업분류코드'] <= 47,'산업별_평균근속기간'] = 5.7
df_period_v2.loc[df_period_v2['산업분류코드'] <= 42,'산업별_평균근속기간'] = 5.7
df_period_v2.loc[df_period_v2['산업분류코드'] <= 39,'산업별_평균근속기간'] = 7.1
df_period_v2.loc[df_period_v2['산업분류코드'] <= 36,'산업별_평균근속기간'] = 13.6
df_period_v2.loc[df_period_v2['산업분류코드'] <= 33,'산업별_평균근속기간'] = 7.7
df_period_v2.loc[df_period_v2['산업분류코드'] <= 8,'산업별_평균근속기간'] = 8.6
df_period_v2.loc[df_period_v2['산업분류코드'] <= 3,'산업별_평균근속기간'] = 8.2
df_period_v2.loc[df_period_v2['산업분류코드'] <= 0,'산업별_평균근속기간'] = 6.7

df_period_v2 = df_period_v2[['사업자등록번호', '산업별_평균근속기간']]


df_period = pd.merge(df_period, df_period_v2, how = 'left', left_on = '사업자등록번호', right_on = '사업자등록번호')
# df_period['종업원_평균근속기간'] = df_period['종업원_평균근속기간'].fillna(7.0)
df_period['평균근속기간'] = (df_period['종업원_평균근속기간'] + df_period['산업별_평균근속기간'])/2
df_period = df_period[['사업자등록번호', '산업분류코드', '평균근속기간']]

df_period
```

```python
df_period_tmp = df_period.groupby('산업분류코드').agg({'평균근속기간':'mean'}).reset_index()
df_period_tmp.columns = ['산업분류코드', '산업별_평균근속기간']

# df_period = pd.merge(df_period, df_period_tmp, how = 'left', left_on = '산업분류코드', right_on = '산업분류코드')
# df_period['평균근속기간'] = np.where(pd.notnull(df_period['산업분류코드']) == True, df_period['평균근속기간'], df_period['산업별_평균근속기간'])

# df_period = df_period[['사업자등록번호', '산업분류코드', '평균근속기간']]
# df_period.columns = ['사업자등록번호', '산업분류코드', '종업원_평균근속기간']


df_period_tmp
```

# **근무환경 (중분류 & 대분류)**

```python
df_tmp = pd.read_csv('/content/drive/Shareddrives/2022 디지털산업혁신 빅데이터 플랫폼 아이디어 경진대회/dataset/기업성장_16종/기업정보상세.csv')
df_infra = df_tmp.copy()

df_infra = df_infra[['사업자등록번호', '산업분류코드']]
df_infra['산업분류코드_v2'] = df_infra['산업분류코드'].fillna(00)

df_infra['산업분류코드_v2'] = df_infra['산업분류코드_v2'].astype('str')
df_infra['산업분류코드_v2'] = df_infra['산업분류코드_v2'].str[:2]
df_infra['산업분류코드_v2'] = df_infra['산업분류코드_v2'].astype('float')


df_infra.loc[df_infra['산업분류코드_v2'] >= 34,'근무환경평균만족도'] = (12.2*5 + 26.4*4 + 41.7*3 + 13.7*2 + 5.3*1)
df_infra.loc[df_infra['산업분류코드_v2'] <= 33,'근무환경평균만족도'] = (7.7*5 + 25.0*4 + 44.4*3 + 15.8*2 + 6.6*1)
df_infra.loc[df_infra['산업분류코드_v2'] <= 3,'근무환경평균만족도'] = (5.6*5 + 10.8*4 + 45.9*3 + 29.3*2 + 6.8*1)
df_infra.loc[df_infra['산업분류코드_v2'] <= 0,'근무환경평균만족도'] = (11.3*5 + 26.0*4 + 42.2*3 + 14.2*2 + 5.6*1)


df_infra = df_infra[['사업자등록번호', '산업분류코드', '근무환경평균만족도']]

df_infra
```
